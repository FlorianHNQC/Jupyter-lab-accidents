```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import matplotlib.cm as cm
%matplotlib inline
f0 = './2022/carcteristiques-2022.csv'
f1 = './2021/carcteristiques-2021.csv'
f2 = './2020/caracteristiques-2020.csv'
f3 = './2019/caracteristiques-2019.csv'
dt0 = pd.read_csv(f0, delimiter=';')
dt1 = pd.read_csv(f1, delimiter=';')
dt2 = pd.read_csv(f2, delimiter=';')
dt3 = pd.read_csv(f3, delimiter=';')

bine = 2

plt.hist(dt0['an'], bins=bine, alpha=1, label='2022')
plt.hist(dt1['an'], bins=bine, alpha=1, label='2021')
plt.hist(dt2['an'], bins=bine, alpha=1, label='2020')
plt.hist(dt3['an'], bins=bine, alpha=1, label='2019')
plt.xlabel('Année')
plt.ylabel('Nb Accidents')
plt.legend(loc='upper right')
plt.title('Histogramme du nombre d\'accidents par année')
plt.xticks([2019, 2020, 2021, 2022])
plt.show()
```
